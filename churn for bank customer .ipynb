# find the customer with credit score between 700 to 850

```python
def credit_score_filter(df):
    df_filter=df[(df['CreditScore']> 700) & (df['CreditScore'] <= 850)]
    x=len(df_filter)
    y=print('No of score more than 700:',x)
    return y,df_filter
```

# find the customer ID that bought num of products:3

```python
def customer_purchase(df):
    df_filter=df[(df['NumOfProducts'])== 3]
    x=len(df_filter)
    y=print('No of customer with no of prodcuts-3:',x)
    customer_detail=df_filter[['CustomerId','CreditScore','NumOfProducts']]
    return y,customer_detail
```

# list of customer ID who left the bank(exited=0)and who not left the bank(exited=1)

```python
def Exited_filter(df):
    Exited_no=df[(df['Exited'] == 1)]
    Exited_no_ID_list=list(Exited_no['CustomerId'])
    Exited_yes=df[(df['Exited'] == 0)]
    Exited_yes_ID_list=list(Exited_yes['CustomerId'])
    return Exited_no_ID_list, Exited_yes_ID_list
    
```

```python
Exited_filter(df)
```

# filter surname startswith 

```python
def surname_filter(df):
    surname_startswith ="H"
    bool_series= df["Surname"].str.startswith(surname_startswith)
    return df[bool_series]
```
